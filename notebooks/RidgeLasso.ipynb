DATA CLEANING

```python
data[data.isnull().any(axis=1)]
```

```python
data.loc[:122,"Region"]=0
data.loc[122:,"Region"]=1
```

```python
data.info()
```

```python
data[['Region']]=data[['Region']].astype('int64')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
## Remove null values
data=data.dropna().reset_index(drop=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.iloc[[122]]
```

```python
data=data.drop(122).reset_index(drop=True)
```

```python
data.iloc[[122]]
```

```python
data.columns
```

```python
## fix spaces in column names
data.columns=[i.strip() for i in data.columns]
```

```python
data.columns
```

```python
data.info()
```

## Change the datatype of some columns


```python
data[['month','day','year','Temperature','RH','Ws']]=data[['month','day','year','Temperature','RH','Ws']].astype('int64')
```

```python
data.info()
```

## Changing the other columns to float64


```python
objects=[features for features in data.columns if data[features].dtype=='object']
```

```python
for i in objects:
    if i!='Classes':
        data[i]=data[i].astype('float64')
```

```python
data.info()
```

```python
data.describe()
```

```python
## let's save the cleaned dataset
data.to_csv('Algerian_forest_fires_dataset_CLEANED.csv',index=False)
```
